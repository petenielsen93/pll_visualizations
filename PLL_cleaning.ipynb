```python
import pandas as pd
import sqlalchemy
```

```python
file = "./Data/pll-player-stats.csv"
pll_df = pd.read_csv(file)
```

```python
pll_df
```

```python
#adding first and last name column, dropping individual columns
pll_df['Name'] = pll_df['First Name'].str.cat(pll_df['Last Name'],sep=" ")
pll_df = pll_df.drop(columns=['First Name', 'Last Name'])
pll_df
```

```python
#reformatting jersey column
pll_df['Jersey'] = pll_df['Jersey'].map("{:.0f}".format)
pll_df.head()
```

```python
#creating df of FOGOs
faceoff_df = pll_df[pll_df['Position'] == 'FO']
faceoff_df = faceoff_df[['Name', 'Position', 'Faceoffs', 'Faceoff Wins', 'Faceoff Losses', 'Faceoff Pct']]
faceoff_df.head()
```

```python
faceoff_df.to_csv('./Data/fogo.csv', index=False)
```

```python
#creating df of Attackmen
attack_df = pll_df[pll_df['Position'] == 'A']
attack_df = attack_df[['Name', 'Position', 'Games Played', 'Total Goals', 'Shots', 'Shot Pct', 'Assists', 'Points']]
attack_df.head()
```

```python
attack_df.to_csv('./Data/attack.csv', index=False)
```

```python
points_df = pll_df.sort_values(by='Points', ascending=False)
points_df.head()
```

```python
points_df = points_df[['Name', 'Position', 'Games Played', 'Total Goals', 'Shots', 'Shot Pct', 'Assists', 'Points']]
points_df
```

```python
point_leaders = points_df.head(10)
point_leaders
```

```python
point_leaders.to_csv('./Data/points.csv', index=False)
```
